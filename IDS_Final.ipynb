```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
#from sklearn import preprocessing
import scipy.stats as sc
```

```python
df = pd.read_csv("33_CH_not_murder_victim_age_sex.csv")
#df.head()
dff=df.fillna(df.median())
dff.head()
#dff.to_csv('Cleaned.csv')
```

```python
plt.scatter(dff['Year'],dff['Victims_Total']) 
plt.title("Number of victims in different states over various years")
plt.xlabel('Year')
plt.ylabel('Total number of Victims')
plt.legend(loc='center')
plt.show()
```

```python
state_list=list(dff.Area_Name)
year=list(set(dff.Year))
total=list(dff.Victims_Total)
group=(list(dff.Sub_Group_Name))
list_mah=[]
for i in range(0,len(state_list)):
    if(state_list[i]=="Maharashtra" and group[i]=="1. Male Victims"):
        list_mah.append(i)
list1=[]
for i in range(0,10):
    list1.append(total[list_mah[i]])
mean_mah=(np.mean(list1))
sd_mah=np.std(list1)
zvalue_mah=(37.7-37.166)/(sd_mah)   
print("P-value of Male Victims in Maharashtra(in 2010) is",sc.norm.cdf(zvalue_mah)-sc.norm.cdf(-1*zvalue_mah))
plt.title('Box plot of Male Victims in Maharashtra(in 2010)')
plt.boxplot(list1)
plt.show()
```
